# EXERCISE 10.1

Questa esercitazione propone una modifica di quanto fatto per l'esercitazione numero 9,ovvero viene richiesto di utilizzare un Simulated Annealing algorithm per risolvere il problema del commesso viaggiatore anzichè utilizzare l'algoritmo Genetico.
Il Simulated Annealing fà comunque appoggio all'algoritmo genetico per quanto riguarda le mutazioni ( ma non il crossover). Il funzionamento è il seguente:
Dopo aver generato le 32 città e costruito i primi percorsi (OldPopulation), utilizzando le mutazioni introdotte nell'esercitazione 9 si costruisce una nuova popolazione di percorsi(NewPopulation). 
Per decidere se queste nuove mutazioni potranno effettivamente far parte della popolazione rispetto a quelle precedenti si utilizza l'algoritmo di Metropolis.
Nel caso in cui il percorso modificato abbia una lunghezza L(x)>L(x)old esso verrà accettato con una probabilità data da $$e^{-1/T*(L(x_{new}-L(x_{old}))}$$, nel caso contrario esso verrà automaticamente accettato in quanto percorso più breve del precedente.
Al tendere della temperatura T verso 0 il sistema convergerà al percorso che minimizza la distanza (in quanto sarà sempre meno probabile accettare il percorso mutato se la sua lunghezza risulta maggiore del percorso prima). 
Per questa simulazione ho variato 100 volte la temperatura $5<T<0.05$, e per ogni temperatura (per ogni blocco) ho fatto 1000 step dell'algoritmo.

Riporto i grafici che rappresentano il percorso migliore e il suo andamento della lunghezza man mano che scendeva la temperatura sia per le città vincolate a una circonferenza che per le città chiuse in un quadrato. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math


pathCirc=np.loadtxt("bestpath_circ.dat")
coord=np.loadtxt("Città_circonferenza.dat")
Lx_min=np.loadtxt("Lx_circ.dat")

T=np.ones(100)
for i in range(100):
    T[i]=5/(i+1)

plt.errorbar(Lx_min[:,0],Lx_min[:,1],Lx_min[:,3])
plt.xlabel("N_blocchi")
plt.ylabel("L(x)")
plt.title("stima di L(x) best path")
plt.grid(True)
plt.show()

#creo i vettori ordinati che minimizzano il percorso L(x) seguendo gli indici caricati in pathCirc(ordino x e y per poterli graficare)
vettX=np.ones(33)
vettY=np.ones(33)

for i in range(32):
    j=int(pathCirc[i])
    vettX[i]=coord[j,1]
    vettY[i]=coord[j,2]
vettX[32]=vettX[0]          #chiudo il vettore facendolo tornare alla città iniziale
vettY[32]=vettY[0]

plt.plot(coord[:,1],coord[:,2],"ro")
plt.plot(vettX,vettY) 
plt.xlabel("x")
plt.ylabel("y")
plt.title("percorso che minimizza L(x)")
plt.grid(True)
plt.show()
```

```python
pathQuad=np.loadtxt("bestpath_quadr.dat")
coord_quad=np.loadtxt("Città_quadrato.dat")
Lx_min_quad=np.loadtxt("Lx_quadr.dat")

plt.errorbar(Lx_min_quad[:,0],Lx_min_quad[:,1],Lx_min_quad[:,3])
plt.xlabel("N_blocchi")
plt.ylabel("L(x)")
plt.title("stima di L(x) best path")
plt.grid(True)
plt.show()

#creo i vettori ordinati che minimizzano il percorso L(x) seguendo gli indici caricati in pathCirc(ordino x e y per poterli graficare)
vettX_q=np.ones(33)
vettY_q=np.ones(33)

for i in range(32):
    j=int(pathQuad[i])
    vettX_q[i]=coord_quad[j,1]
    vettY_q[i]=coord_quad[j,2]
vettX_q[32]=vettX_q[0]           #chiudo il vettore facendolo tornare alla città iniziale
vettY_q[32]=vettY_q[0]

plt.plot(coord_quad[:,1],coord_quad[:,2],"ro")
plt.plot([0,2,2,0,0],[0,0,2,2,0],"b--")
plt.plot(vettX_q,vettY_q) 
plt.xlabel("x")
plt.ylabel("y")
plt.title("percorso che minimizza L(x)")
plt.grid(True)
plt.show()
```
