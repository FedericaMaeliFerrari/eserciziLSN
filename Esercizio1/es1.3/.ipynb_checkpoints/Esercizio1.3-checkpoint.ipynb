# EXERCISE 1.3

l'esercizio consiste nella simulazione dell'esperimento di Buffon. esso consiste nel lancio di un legnetto (da me fissato a lunghezza $L=2 cm$) su un tavolo segnato da linee orizzontali a distanza d (fissata a $d=10 cm$). Questo esperimento può essere utilizzato per valutare il valore di $\pi$: dati $N_{lanci}$, di cui $N_{hit}$ colpiscono le righe orizzontali, $\pi$ può essere stimato come $$\pi = \frac{2L}{Pd} = \lim_{N_{lanci} \to \infty}\frac{2LN_{lanci}}{N_{hit}d}$$
Non potendo utilizzare il valore di $\pi$ per generare l'inclinazione di caduta del legnetto (che seguirebbe la distribuzione seno), ho utilizzato il metodo ACCEPT-REJECT. Dopo aver generato la posizione di caduta del baricentro del legnetto sul tavolo, ho generato le coordinate $x$,$y$ di uno dei due vertici rispetto al baricentro all'interno del quadrato di lato $L/2$ : se il punto così generato si trovava all'interno della circonferenza di raggio $L/2$ il dato veniva considerato buono e, riportandosi sulla circonferenza, veniva calcolata la distanza effettiva rispetto all'asse $y$ tra il baricentro e l'estremo.

Come si può osservare dal grafico sottostante all'aumentare degli $N_{lanci}$ considerati per la media progressiva il valore di $\pi$ calcolato converge al valore reale di $\pi$ (linea arancione sul grafico), giustificando l'esattezza del metodo. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

N=100
L=1000
pival,err=np.loadtxt("DatiPi.dat",usecols=(0,1),unpack=True)#carico dal file Dati.dat i vettori
x=np.arange(N)*L
p=np.ones(N)*math.pi
#grafico valori Pi calcolati al variare di N.
plt.errorbar(x,pival,err)
plt.plot(x,p)
plt.xlabel('#throws')
plt.ylabel('$\pi$')
plt.title("Grafico valore di $\pi$")
plt.grid(True)
print("Valore reale pi greco:")
print(math.pi)
print("Valore a cui converge pi greco calcolato:")
print(pival[N-1])
plt.show()
```
