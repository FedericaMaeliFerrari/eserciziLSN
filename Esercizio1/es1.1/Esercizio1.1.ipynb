# EXERCISE 1.1-1 e 1.1-2
1.  $\langle r \rangle = \int_0^1 r dr = 1/2$
2. $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$ 

Lo scopo dell'esercizio è valutare $\langle r \rangle$ e $\sigma^2$ con la loro relativa incertezza (Deviazione Standard) al variare dei blocchi considerati per il calcolo. Il programma prende in ingresso un file di dati (vedere es1.1.cc e classe Random) contenente i vettori.Vengono dunque plottati in due grafici la stima di $\langle r \rangle$ e di $\sigmaì2$ con le relative barre di errore. Come si ci aspettava, all'aumentare del numero dei valori considerati per il calcolo sia l'integrale 1 che l'integrale 2 convergono a zero,confermando il corretto funzionamento del Pseudo Random Number

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

N=100
L=1000
y,err,sigma,sigma_err=np.loadtxt("Dati.dat",usecols=(0,1,2,3),unpack=True)#carico dal file Dati.dat i vettori
x=np.arange(N)*L
r_att=np.ones(100)*0

#grafico medie <r>
plt.errorbar(x,y-0.5,err)
plt.plot(x,r_att,label="Valore atteso")
plt.xlabel('#throws')
plt.ylabel('$<r>-1/2$')
plt.title("Grafico 1_<r>")
plt.grid(True)
plt.show()

#grafico sigma^2
plt.figure()
plt.errorbar(x,sigma-1/12,sigma_err)
plt.plot(x,r_att,label="Valore atteso")
plt.legend()
plt.xlabel('#throws')
plt.ylabel('$<(r-0.5)^2>-1/12$')
plt.title("Grafico 2_$\sigma^2$")
plt.grid(True)
plt.show()
    

```

# EXERCISE 1.1-3

Lo scopo dell'esercizio è eseguire il test del $\chi^2$ su $10^4$ numeri generati tra $[0,1]$ per $100$ volte.
L'intervallo $[0,1]$ viene dunque suddiviso a sua volta in $100$ intervalli M, e si tiene conto di quanti numeri cadono in ciascun intervallo. Il $\chi^2$ viene calcolato dunque come
$$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}$$
con $n_i$ quantità di numeri compresi nell'intervallo i-esimo, $n/M$ probabilità attesa, ovvero numero di lanci totali su numero di sub-intervalli ($ n/M = 10^4/100 =100$). Questa operazione viene svolta per 100 ripetizioni, ottenendo un vettore $\chi^2_j$ per $j=1,...100$ (preso in ingresso dal programma da un file di dati "DatiChi.dat", vedere es.1.1_c.cc e classe Random). 
I valori del $\chi^2$ oscillano intorno a $100$, in accordo alla distribuzione uniforme $np = n\times1/M$, confermando l'esattezza del test.




```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math


y=np.loadtxt("DatiChi.dat",usecols=(0),unpack=True)
X=len(y)
x=np.arange(X)
att=np.ones(100)*100

plt.plot(x,y)
plt.plot(x,att,label="Valore atteso")
plt.legend()
plt.xlabel("NumSimul")
plt.ylabel("$\chi^2$")
plt.grid(True)
plt.show()
```
