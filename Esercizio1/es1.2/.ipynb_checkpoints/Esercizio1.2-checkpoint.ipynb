# EXERCISE 1.2
Per l'esercizio sono state condotte tre simulazioni, generando numeri casuali secono la distribuzione uniforme, la distribuzione esponenziale e la distribuzione lorentziana. Per ogni distribuzione sono stati generati $10^4$ valori di $S_n= \frac{1}{N}\sum_{i=1}^N x_i$ (per $N=1, 2, 10, 100$), con $x_i$ variabile generata secondo le tre distribuzioni. Per la distribuzione esponenziale $p(x) = \lambda \exp(-\lambda x)$, $x\in [0;+\infty]$ si è tenuto fisso il valore di $\lambda=1$, mentre per la distribuzione lorentziana $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$, $x\in [-\infty;+\infty]$ si sono fissati i parametri $\mu=0$ e $\Gamma=1$.


Come si osserva dai grafici sottostanti, nel caso della distribuzione uniforme all'aumentare di $N$ l'istogramma assume un profilo gaussiano, stesso discorso per l'andamento esponenziale.Per i numeri generati secondo la distribuzione lorentziana vale un discorso diverso, in quanto essa mostra per tutti gli $N$ il tipico andamento della funzione di Lorentz: non tutte le distribuzioni forniscono medie che convergono alla distribuzione normale, ovvero non per tutte le distribuzioni vale il teorema del Limite Centrale

NB l'ultimo fit per la distribuzione lorentziana pare non corretto perchè l'istogramma riportato è frutto di uno "zoom" tra i valori $[-50,50]$. Non sono riuscita a fare lo stesso per il fit, che dunque fitta la distribuzione la curva data da tutti i numeri generati, e da ciò nasce la discrepanza. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit

N=np.array(["1","2","10","100"])
y=np.loadtxt("Dati[0,1].dat")#carico dal file Dati[0,1].dat i vettori riguardanti la distrib uniforme
ye=np.loadtxt("DatiExp.dat")#carico dal file DatiExp.dat i vettori riguardanti la distrib esponenziale
yl=np.loadtxt("DatiLorentz.dat")#carico dal file DatiLorentz.dat i vettori riguardanti la distrib lorentziana

def fgauss(x,norm,mu,sigma):  # definisco funzione di fit gaussiana
    return (norm/(sigma*np.sqrt(2*math.pi)))* np.exp(-(x-mu)**2/(2*sigma**2))


for i in range(4):
    n,bins,columns=plt.hist(y[:,i],bins="auto")
    plt.ylabel("n")
    plt.xlabel("x")
    plt.title("DISTRIBUZIONE UNIFORME (N="+N[i%4]+") ")
    x=np.zeros(len(bins)-1)
    if i==3:
        for j in range(len(bins)-1):
            x[j]=(bins[j]+bins[j+1])/2
        p_opt, p_cov = curve_fit(fgauss, x, n)
        print("Parametri ottimizzati [norm,mu,sigma]:")
        print(p_opt)
        plt.plot(x,fgauss(x,p_opt[0],p_opt[1],p_opt[2]))
    plt.grid(True)
    plt.show()
    
```

```python
for i in range(4):
    n,bins,columns=plt.hist(ye[:,i],bins="auto")
    plt.ylabel("n")
    plt.xlabel("x")
    plt.title("DISTRIBUZIONE ESPONENZIALE (N="+N[i%4]+") ")
    x=np.zeros(len(bins)-1)
    if i==3:
        for j in range(len(bins)-1):
            x[j]=(bins[j]+bins[j+1])/2
        p_opt, p_cov = curve_fit(fgauss, x, n)
        print("Parametri ottimizzati [norm,mu,sigma]:")
        print(p_opt)
        plt.plot(x,fgauss(x,p_opt[0],p_opt[1],p_opt[2]))
    plt.grid(True)
    plt.show()
```

```python
def florentz(x,mu,G):      #definisco funzione di fit lorentziana
    return (1/math.pi)*G/((x-mu)**2+G**2)

for i in range(4):
    n,bins,columns=plt.hist(yl[:,i],bins="auto",range=(-50,50))
    plt.ylabel("n")
    plt.xlabel("x")
    plt.title("DISTRIBUZIONE LORENTZIANA (N="+N[i%4]+") ")
    x=np.zeros(len(bins)-1)
    if i==3:
        for j in range(len(bins)-1):
            x[j]=(bins[j]+bins[j+1])/2
        p_opt, p_cov = curve_fit(florentz, x, n)
        print("Parametri ottimizzati [norm,mu,sigma]:")
        print(p_opt)
        plt.plot(x,florentz(x,p_opt[0],p_opt[1]))
    plt.grid(True)
    plt.show()
```
