### Exercise 4.1
Lo scopo dell'esercizio è quello di simulare il moto di 108 molecole contenute (in questo caso) in un cubo allo stato liquido. Le ipotesi assunte alla base della simulazione sono:

1) Pur essendo all'interno di un contenitore si trascurano gli effetti di superficie, immaginando di replicare in modo periodico le condizioni al contorno-->costruzione di un reticolo infinito 

2) All'interno del cubo la densità rimane invariata, dunque qualora una molecola uscisse dal cubo subito ne rientrerebbe un'altra

3) Le molecole sono soggette al potenziale di Lennard-Jones $v(r) =4[(1/r)^{12}-(1/r)^6]$ entro un certo raggio d'azione $r_{cut}$

4) Tutte le quantità sono espresse in unità ridotte

L'algoritmo usato per risolvere le equazioni del moto è l'algoritmo di Velvet $$\vec{r}(t+dt)=2\vec{r}(t)-\vec{r}(t-dt)+dt^2\vec{a}(t)$$

Per questa simulazione ho utilizzato l'opzione Restart, in cui sono state caricate le posizioni al tempo (t) e al tempo (t-dt). La velocità dunque (calcolata come $\vec{v}(t+dt)=\frac{\vec{r}(t+dt)-\vec{r}(t)}{dt}$) è stata riscalata per ottenere la temperatura desiderata, correggendo dunque le posizioni $\vec{r}(t-dt)$

Ho eseguito dunque dieci simulazioni con l'opzione Restart, e in ognuna di essere ho eseguito 1000 step. Dal grafico si può osservare come il sistema già all'inizio della simulazione abbia praticamente raggiunto l'equilibrio (in quanto prima dell'opzione Restart il sistema è stato fatto evolvere per 10000 step usando delle velocità randomiche). Sopratturro dal grafico dell'Energia totale si può osservare la stabilità del sistema all'interno di ogni simulazione. Man mano che si va avanti con le simulazioni il salto a gradino diviene via via più piccolo, in quanto il sistema diventa sempre più stabile.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit

Temp=np.loadtxt("Temperatura_Restart.dat")
Ekin=np.loadtxt("EnCinetica_Restart.dat")
Etot=np.loadtxt("EnergiaTot_Restart.dat")
Epot=np.loadtxt("Potenziale_Restart.dat")
num=int(len(Temp))
x=np.arange(num)

plt.plot(x,Temp,label="T")
plt.plot(x,Ekin,label="Ekin")
plt.plot(x,Epot,label="Epot")
plt.plot(x,Etot,label="Etot")
plt.xlabel("Nstep")
plt.ylabel("Valori simulazione")
plt.legend()
plt.grid(True)
plt.show()


plt.plot(x,Etot,label="Etot")
plt.xlabel("Nstep")
plt.ylabel("Etot")
plt.title("Focus sull'energia totale del sistema")
plt.legend()
plt.grid(True)
plt.show()

plt.plot(x,Temp,label="T")
plt.xlabel("Nstep")
plt.ylabel("T")
plt.title("Focus sulla temperatura interna del sistema")
plt.legend()
plt.grid(True)
plt.show()
```
