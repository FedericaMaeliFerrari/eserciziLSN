```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit

simulTemp=np.loadtxt("output.Temp.0")
simulEpot=np.loadtxt("output.Epot.0")
simulEkin=np.loadtxt("output.Ekin.0")
simulEtot=np.loadtxt("output.Etot.0")
x=np.arange(100)*1000
epskb=120
eps=120*1.38*pow(10,-23)


#Temperatura
plt.errorbar(x,simulTemp[0:100,2]*epskb,simulTemp[0:100,3]*epskb,label="Gas")     #gas 
plt.errorbar(x,simulTemp[100:200,2]*epskb,simulTemp[100:200,3]*epskb,label="Liquido") #liquido
plt.errorbar(x,simulTemp[200:300,2]*epskb,simulTemp[200:300,3]*epskb,label="Solido") #solido
plt.xlabel("Nstep")
plt.ylabel("T")
plt.title("Stima di T")
plt.legend()
plt.grid(True)
plt.show()

#Energia cinetica
plt.errorbar(x,simulEkin[0:100,2]*eps,simulEkin[0:100,3]*eps,label="Gas")     #gas 
plt.errorbar(x,simulEkin[100:200,2]*eps,simulEkin[100:200,3]*eps,label="Liquido") #liquido
plt.errorbar(x,simulEkin[200:300,2]*eps,simulEkin[200:300,3]*eps,label="Solido") #solido
plt.xlabel("Nstep")
plt.ylabel("Ekin")
plt.title("Stima di Ekin")
plt.legend()
plt.grid(True)
plt.show()


#Energia potenziale
plt.errorbar(x,simulEpot[0:100,2]*eps,simulEpot[0:100,3]*eps,label="Gas")     #gas 
plt.errorbar(x,simulEpot[100:200,2]*eps,simulEpot[100:200,3]*eps,label="Liquido") #liquido
plt.errorbar(x,simulEpot[200:300,2]*eps,simulEpot[200:300,3]*eps,label="Solido") #solido
plt.xlabel("Nstep")
plt.ylabel("Epot")
plt.title("Stima di Epot")
plt.legend()
plt.grid(True)
plt.show()


#Energia totale
plt.errorbar(simulEtot[0:100,2]*eps,simulEtot[0:100,3]*eps,label="Gas")     #gas 
plt.errorbar(x,simulEtot[100:200,2]*eps,simulEtot[100:200,3]*eps,label="Liquido") #liquido
plt.errorbar(x,simulEtot[200:300,2]*eps,simulEtot[200:300,3]*eps,label="Solido") #solido
plt.xlabel("Nstep")
plt.ylabel("Etot")
plt.title("Stima di Etot")
plt.legend()
plt.grid(True)
plt.show()
```
