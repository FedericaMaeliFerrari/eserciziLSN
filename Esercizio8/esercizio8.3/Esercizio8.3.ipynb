# EXERCISE 8.2

In questa parte si richiede di utilizzare il codice PIM/PIGS fornitoci per studiare lo stessa sistema quantistico.
Inizialmente l'algoritmo viene utilizzato per $T=0$ (ground-state) per una funzione d'onda costante $\Psi_T(x)=1$.
Sono riportati i grafici con l'andamento dell'energia cinetica e potenziale e il profilo del comportamento della funzione d'onda al ground-state.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit

Ekin=np.loadtxt("kinetic.dat")
V=np.loadtxt("potential.dat")
px=np.loadtxt("probability.dat")

plt.errorbar(Ekin[:,0],Ekin[:,1],Ekin[:,2],label="Energia cinetica")
plt.errorbar(V[:,0],V[:,1],V[:,2],label="Potenziale")
plt.xlabel("timeslices")
plt.title("Andamento energia e potenziale $\psi =1$")
plt.grid(True)
plt.legend()
plt.show()

plt.errorbar(px[:,0],px[:,1],px[:,2])
plt.xlabel("x")
plt.ylabel("$\psi$")
plt.title("Ground state della funzione d'onda costante $\psi =1$")
plt.grid(True)
plt.show()
```

Successivamente l'algoritmo viene fatto andare con $\Psi_T(x) = \Psi_T^{\sigma,\mu}(x)$, funzione d'onda utilizzata nella parte precedente con $\mu$ e $\sigma$ aventi i valori ottimizzati nella parte $8.2$ ($\mu=\sigma=0.8$)
Al solito vengono riportati i grafici dell'energia cinetica e del potenziale e il grafico con l'andamento della funzione d'onda al ground state, confrontato con quello ricavato dall'esercitazione 8.2.


```python
Ekin=np.loadtxt("kinetic.dat")
V=np.loadtxt("potential.dat")
px=np.loadtxt("probability.dat")
pxmis=np.loadtxt("posX.dat")

plt.errorbar(Ekin[:,0],Ekin[:,1],Ekin[:,2],label="Energia cinetica")
plt.errorbar(V[:,0],V[:,1],V[:,2],label="Potenziale")
plt.xlabel("timeslices")
plt.title("Andamento energia e potenziale $\Psi_T^{\sigma,\mu}$")
plt.grid(True)
plt.legend()
plt.show()

plt.errorbar(V[:,0],V[:,1],V[:,2],label="Potenziale")
plt.xlabel("timeslices")
plt.ylabel("V")
plt.title("Focus potenziale")
plt.grid(True)
plt.legend()
plt.show()

plt.errorbar(px[:,0],px[:,1],px[:,2],label="quantum code")
plt.hist(pxmis, bins=100, range=(-2, 2), density=True, stacked=True,label="VMC code")
plt.xlabel("x")
plt.ylabel("$\psi$")
plt.title("Ground state della funzione d'onda costante $\Psi_T^{\sigma,\mu}$")
plt.legend()
plt.grid(True)
plt.show()
```

Infine la simulazione viene fatta andare aumentando la temperatura T. 
Nel grafico sono riportati gli andamenti della funzione d'onda al variare della temperatura.
Quello che si osserva è una diminuzione dell'energia della funzione d'onda all'aumentare della temperatura T dovuta a un allargamento della stessa.

```python
p_0=np.loadtxt("probability_0.dat")
p_1=np.loadtxt("probability_1.25.dat")
p_2=np.loadtxt("probability_10.dat")
p_3=np.loadtxt("probability_20.dat")

plt.plot(p_0[:,0],p_0[:,1],label="ground state")
plt.plot(p_1[:,0],p_1[:,1],label="T=1.25")
plt.plot(p_2[:,0],p_2[:,1],label="T=10")
plt.plot(p_3[:,0],p_3[:,1],label="T=20")
plt.xlabel("x")
plt.ylabel("$\psi$")
plt.title("Andamento della funzione d'onda all'aumentare della temperatura")
plt.legend()
plt.grid(True)
plt.show()
```
