La seconda parte dell'esercitazione chiede di trovare i parametri $\mu$ e $\sigma$ che minimizzano $\langle {\hat H} \rangle_T$:

$$
\langle {\hat H} \rangle_T = \frac{\langle \Psi_T^{\sigma,\mu}| {\hat H} | \Psi_T^{\sigma,\mu} \rangle}
{\langle \Psi_T^{\sigma,\mu} | \Psi_T^{\sigma,\mu} \rangle} \ge E_0 =
\frac{\langle \Psi_0| {\hat H} | \Psi_0 \rangle}
{\langle \Psi_0 | \Psi_0 \rangle}
$$

I parametri torvati che minimizzano la funzione sono $\mu=0.8$ e $\sigma=0.8$, con un delta per l'algoritmo di Metropolis pari a $\delta=2.95$.
Il primo grafico mostra l'andamento di  $\langle {\hat H} \rangle_T$ campionato con questi parametri.

Oltre a ciò viene richiesto di riempire un istogramma con le configurazioni campionate dall'algoritmo di Metropolis in modo da mostrare l'andamento della funzione $|\Psi_T(x)|^2$ comparato alla sua soluzione analitica.
La soluzione analitica viene calcolata trasformando l'equazione di Schrodinger in una equazione matriciale:
gli elementi diagonali della matrice possono essere interpretati come distribuzioni di densità nelle coordinate spaziali, mentre gli elementi non diagonali come probabilità di transione da una posizione $x$ a una posizione $x'$.
Ciò viene mostrato nel secondo grafico.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit

def Vpot(x):
    return (x**2 - 2.5)*x**2


H=np.loadtxt("stima_H.dat",skiprows=1)  #salto il primo blocco per cominciare la simulazione con dati già equilibrati
x=np.loadtxt("posX.dat")

plt.errorbar(H[:,0]-1,H[:,2],H[:,3])
plt.plot(H[:,0]-1,np.ones(100)*H[99,2],label="-0.454726")
plt.xlabel("N_blocchi")
plt.ylabel("<H>")
plt.title("stima del valor medio di H")
plt.grid(True)
plt.legend()
plt.show()

plt.hist(x, bins=100, range=(-2, 2), density=True, stacked=True)  #istogramma dei valori di x campionati durante la simulazione

#SOLUZIONE MATRICIALE
hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)
scale = 0.3
plt.plot(x,((psi[0])**2))
plt.xlabel("x")
plt.grid(True)
plt.xlim((-2,2))
plt.show()
```

Pur non essendo l'istogramma fittato bene dall'andamento atteso di $|\Psi_T(x)|^2$ si possono riconoscere al suo interno la sovrapposizione di due gaussiane. La discrepanza è data dal fatto che quello trovato non corrisponde al vero valor minimo dell'energia, in quanto la funzione d'onda da noi utilizzata non è altro che una approssimazione della funzione d'onda al ground state (che non corrsiponde infatti alla somma di due gaussiane). Inoltre i paretri trovati $\mu$ e $\sigma$ che minimizzano $\langle {\hat H} \rangle_T$ possiedono un certo errore, in quanto per trovarli non ho fatto altro che cambiarli di volta in volta e ho scelto quelli che portavano ad avere un valore di $\langle {\hat H} \rangle_T$ minore---> mi trovo sicuramente molto vicino al punto di minimo, ma non è detto che mi trovi nel minimo stesso.
