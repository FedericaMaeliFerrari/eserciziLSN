Il calcolo della distribuzione radiale $g(r)$ per lo stato gassoso,liquido e solido Ã¨ stata fatta anche per un sistema microcanonico (sistema utilizzato per la esercitazione 4, in cui le equazioni del moto erano state risolte utilizzando l'algoritmo di Velvet).
I grafici mostrano come i valori della distribuzione radiale per l'insieme microcanonico siano solo smorzati rispetto a quelli per l'insieme canonico, preservandone l'andamento.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit

ekb=120
e=120*1.38*pow(10,-23)
sigma=0.34*pow(10,-9)

Gr_gas_can=np.loadtxt("output.gave.0",max_rows=100)              #g(r)_gas_canonico
Gr_liq_can=np.loadtxt("output.gave.0",skiprows=100,max_rows=100)
Gr_sol_can=np.loadtxt("output.gave.0",skiprows=200)

Gr_gas_microcan=np.loadtxt("output.gave.4",max_rows=100)              #g(r)_gas_microcanonico
Gr_liq_microcan=np.loadtxt("output.gave.4",skiprows=100,max_rows=100)
Gr_sol_microcan=np.loadtxt("output.gave.4",skiprows=200)


#G(R) a confronto tra insieme canonico e non canonico
plt.plot(Gr_gas_can[:,0]*sigma,Gr_gas_can[:,1],"-bo",label="canonico")
plt.plot(Gr_gas_microcan[:,0]*sigma,Gr_gas_microcan[:,1],"-ro",label="microcanonico")
plt.xlabel("$r/\sigma$")
plt.ylabel("$g(r)$")
plt.title("Distribuzione radiale Gas")
plt.legend()
plt.grid(True)
plt.show()

plt.plot(Gr_liq_can[:,0]*sigma,Gr_liq_can[:,1],"-bo",label="canonico")
plt.plot(Gr_liq_microcan[:,0]*sigma,Gr_liq_microcan[:,1],"-ro",label="microcanonico")
plt.xlabel("$r/\sigma$")
plt.ylabel("$g(r)$")
plt.title("Distribuzione radiale Liquido")
plt.legend()
plt.grid(True)
plt.show()

plt.plot(Gr_sol_can[:,0]*sigma,Gr_sol_can[:,1],"-bo",label="canonico")
plt.plot(Gr_sol_microcan[:,0]*sigma,Gr_sol_microcan[:,1],"-ro",label="microcanonico")
plt.xlabel("$r/\sigma$")
plt.ylabel("$g(r)$")
plt.title("Distribuzione radiale Solido")
plt.legend()
plt.grid(True)
plt.show()
```
