```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit

ekb=120
e=120*1.38*pow(10,-23)
sigma=0.34*pow(10,-9)

Gr_gas_can=np.loadtxt("output.gave.0",max_rows=100)              #g(r)_gas_canonico
Gr_liq_can=np.loadtxt("output.gave.0",skiprows=100,max_rows=100)
Gr_sol_can=np.loadtxt("output.gave.0",skiprows=200)

Gr_gas_microcan=np.loadtxt("output.gave.4",max_rows=100)              #g(r)_gas_microcanonico
Gr_liq_microcan=np.loadtxt("output.gave.4",skiprows=100,max_rows=100)
Gr_sol_microcan=np.loadtxt("output.gave.4",skiprows=200)


en_gas_can=np.loadtxt("output.epot.0",max_rows=1000)              #energia potenziale_gas_canonico
en_liq_can=np.loadtxt("output.epot.0",skiprows=1000,max_rows=1000)
en_sol_can=np.loadtxt("output.epot.0",skiprows=2000)

en_gas_microcan=np.loadtxt("output.epot.4",max_rows=100)              #energia potenziale_gas_microcanonico
en_liq_microcan=np.loadtxt("output.epot.4",skiprows=100,max_rows=100)
en_sol_microcan=np.loadtxt("output.epot.4",skiprows=200)


p_gas_can=np.loadtxt("output.pres.0",max_rows=1000)              #pressione_gas_canonico
p_liq_can=np.loadtxt("output.pres.0",skiprows=1000,max_rows=1000)
p_sol_can=np.loadtxt("output.pres.0",skiprows=2000)

p_gas_microcan=np.loadtxt("output.pres.4",max_rows=100)              #pressione_gas_microcanonico
p_liq_microcan=np.loadtxt("output.pres.4",skiprows=100,max_rows=100)
p_sol_microcan=np.loadtxt("output.pres.4",skiprows=200)


#G(R) a confronto tra insieme canonico e non canonico
plt.plot(Gr_gas_can[:,0]*sigma,Gr_gas_can[:,1],"-bo",label="canonico")
plt.plot(Gr_gas_microcan[:,0]*sigma,Gr_gas_microcan[:,1],"-ro",label="microcanonico")
plt.xlabel("$r/\sigma$")
plt.ylabel("$g(r)$")
plt.title("Distribuzione radiale Gas")
plt.legend()
plt.grid(True)
plt.show()

plt.plot(Gr_liq_can[:,0]*sigma,Gr_liq_can[:,1],"-bo",label="canonico")
plt.plot(Gr_liq_microcan[:,0]*sigma,Gr_liq_microcan[:,1],"-ro",label="microcanonico")
plt.xlabel("$r/\sigma$")
plt.ylabel("$g(r)$")
plt.title("Distribuzione radiale Liquido")
plt.legend()
plt.grid(True)
plt.show()

plt.plot(Gr_sol_can[:,0]*sigma,Gr_sol_can[:,1],"-bo",label="canonico")
plt.plot(Gr_sol_microcan[:,0]*sigma,Gr_sol_microcan[:,1],"-ro",label="microcanonico")
plt.xlabel("$r/\sigma$")
plt.ylabel("$g(r)$")
plt.title("Distribuzione radiale Solido")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(en_gas_can[:,0],en_gas_can[:,2],label="canonico")
plt.plot(en_gas_microcan[:,0]*10,en_gas_microcan[:,2],label="microcanonico")
plt.xlabel("$num blocchi$")
plt.ylabel("$Epot$")
plt.title("Energia potenziale Gas")
plt.legend()
plt.grid(True)
plt.show()

plt.plot(en_liq_can[:,0],en_liq_can[:,2],label="canonico")
plt.plot(en_liq_microcan[:,0]*10,en_liq_microcan[:,2],label="microcanonico")
plt.xlabel("$num blocchi$")
plt.ylabel("$Epot$")
plt.title("Energia potenziale Liquido")
plt.legend()
plt.grid(True)
plt.show()

plt.plot(en_sol_can[:,0],en_sol_can[:,2],label="canonico")
plt.plot(en_sol_microcan[:,0]*10,en_sol_microcan[:,2],label="microcanonico")
plt.xlabel("$num blocchi$")
plt.ylabel("$Epot$")
plt.title("Energia potenziale Solido")
plt.legend()
plt.grid(True)
plt.show()
```
