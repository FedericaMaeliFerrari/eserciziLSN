La quantità di informazione riportata in questi grafici è ridondante. Per avere una buona stima di U/N e P basta prendere le misure dopo un tempo $\tau_c$, tempo in cui il sistema giunge a uno stato significativamente diverso dal primo. Infatti misurare ad ogni step U/N e P porta ad avere set di valori troppo identici, con dunque un alto livello di autocorrelazione.
Per valutare il tempo $\tau_c$, ovvero il tempo che ci impiega l'autocorrelazione a cadere giù verso lo zero (ovvero quando il processo diventa statisticamente indipendente), studio l'andamento della funzione

$$\chi(t)=\frac{\frac{1}{t_{max}-t}\sum_{t'=0}^{t_{max}-t}x(t')x(t'+t)-\frac{1}{t_{max}-t}\sum_{t'=0}^{t_{max}-t}(x(t'))\frac{1}{t_{max}-t}\sum_{t'=0}^{t_{max}-t}(x(t'+t))}{\frac{1}{t_{max}}\sum_{t'=0}^{t_{max}}x(t')^2-(\frac{1}{t_{max}}\sum_{t'=0}^{t_{max}}x(t'))^2}$$


Con $t_{max}$ tempo totale della simulazione (che corrisponde, scegliendo per ipotesi che la simulazione ci mette un secondo a compiere uno step, alla lunghezza del vettore). 
Sotto sono riportati i grafici dell'autocorrelazione per tutte e tre le simulazioni, con $0<=t<=1000$.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit
    
def autocorr(x,passo,dim,dev):
    global prod
    global somma_parz
    global somma_parz2
    prod=0
    somma_parz=0
    somma_parz2=0
    
    for i in range(dim-passo):
        prod+=x[i]*x[i+passo]
        somma_parz+=x[i]
        somma_parz2+=x[i+passo]
    
    prod/=(dim-passo)
    somma_parz/=(dim-passo)
    somma_parz2/=(dim-passo)

    return (prod-somma_parz*somma_parz2)/dev

def devstd(x,dim): 
    global somma
    global somma2
    somma=0
    somma2=0
    for i in range(dim):
        somma+=x[i]
        somma2+=pow(x[i],2)
    
    return (somma2/dim-pow(somma/dim,2))


#TOLGO I PRIMI 15000 VALORI , IN MODO DA ESSERE ALL'EQUILIBRIO
gas=np.loadtxt("Gas.dat",skiprows=1000)
liq=np.loadtxt("Liquido.dat",skiprows=1000)
sol=np.loadtxt("Solido.dat",skiprows=1000)
dim=int(len(gas[:,0]))


#autocorrelazione con il passo t[0:1000]
t=np.arange(1000)
corr_u=np.zeros(1000)
corr_p=np.zeros(1000)
corr_ul=np.zeros(1000)
corr_pl=np.zeros(1000)
corr_us=np.zeros(1000)
corr_ps=np.zeros(1000)

dev_u=devstd(gas[:,0],dim)
dev_p=devstd(gas[:,1],dim)
dev_ul=devstd(liq[:,0],dim)
dev_pl=devstd(liq[:,1],dim)
dev_us=devstd(sol[:,0],dim)
dev_ps=devstd(sol[:,1],dim)

for i in range(1000):
    corr_u[i]=autocorr(gas[:,0],i,dim,dev_u)
    corr_p[i]=autocorr(gas[:,1],i,dim,dev_p)
    corr_ul[i]=autocorr(liq[:,0],i,dim,dev_ul)
    corr_pl[i]=autocorr(liq[:,1],i,dim,dev_pl)
    corr_us[i]=autocorr(sol[:,0],i,dim,dev_us)
    corr_ps[i]=autocorr(sol[:,1],i,dim,dev_ps)

plt.plot(t,corr_u,label="autocorrelazione U/N")
plt.plot(t,corr_p,label="autocorrelazione P")
plt.xlabel("t")
plt.ylabel("$\chi (t)$")
plt.title("Autocorrelazione gas")
plt.legend()
plt.grid(True)
plt.show()

plt.plot(t,corr_ul,label="autocorrelazione U/N")
plt.plot(t,corr_pl,label="autocorrelazione P")
plt.xlabel("t")
plt.ylabel("$\chi (t)$")
plt.title("Autocorrelazione liquido")
plt.legend()
plt.grid(True)
plt.show()
        
plt.plot(t,corr_us,label="autocorrelazione U/N")
plt.plot(t,corr_ps,label="autocorrelazione P")
plt.xlabel("t")
plt.ylabel("$\chi (t)$")
plt.title("Autocorrelazione solido")
plt.legend()
plt.grid(True)
plt.show()
      
```

Come si può osservare dai grafici l'autocorrelazione $\chi(t)$ segue un andamento esponenziale (più op meno ripido a seconda se le particelle sono state simulate alla fase gassosa, liquida, solida).
Già dopo 100 step l'autocorrelazione in tutti i casi cade drasticamente verso zero, assicurandomi dunque che si campionando U/N e P dopo questo lasso di tempo otterei delle misure scorrelate.
Conoscere quanto vale $\tau_c$ permette di stimare quanto deve essere largo un blocco L per il calcolo della media a blocchi. Il blocco deve essere infatti abbastanza largo per far sì che i valori medi di U/N e P siano tra loro scorrelati (e dunque dallo studio dell'autocorrelazione condotto si evince che un blocco largo da 100 a 200 dia una buona stima).

Per controprova studio l'andamento dell'errore statistico della media a blocchi considerando blocchi di larghezza L compresa tra 10 e 5000. Considero per il calcolo solo le L che appartengono all'insieme dei divisori del numero di step M (lunghezza del vettore), in modo da ottenere un numero di blocchi N intero.
In questo caso, per equilibrare il sistema, tolgo i primi 15000 punti, in modo da avere abbastanza L per osservare un andamento definito dell'errore statistico.


```python
def devstandard(M,l,x):
    N=int(M/l)
    m=0
    m2=0
    media=np.zeros(N)
    media2=np.zeros(N)
    for i in range(N):                ##calcolo vettore medie degli N blocchi larghi l
        for j in range(l):
            media[i]+=x[i*l+j]
        media[i]/=l
        media2[i]=(media[i])**2
    for k in range(N):               ##calcolo errore associato alla somma di tutte le N medie (mi serve solo il valore finale)
        m+=media[k]
        m2+=media2[k]
    m/=N
    m2/=N
    return math.sqrt((m2-pow(m,2))/(N-1)) 
        
        


#carico vettori dai files di dati
gas=np.loadtxt("Gas.dat",skiprows=15000)
liq=np.loadtxt("Liquido.dat",skiprows=15000)
sol=np.loadtxt("Solido.dat",skiprows=15000)

#vedo quanti divisori ho di M da L=10 a L=5000
M=int(len(gas[:,0]))
cont=0
for l in range(10,5001):
    if M%l==0:
        cont+=1
        
#creo vettori 
L=np.zeros(cont)
Gas_errU=np.zeros(cont)
Gas_errP=np.zeros(cont)
Liq_errU=np.zeros(cont)
Liq_errP=np.zeros(cont)
Sol_errU=np.zeros(cont)
Sol_errP=np.zeros(cont)

#carico il vettore L(larghezza blocchi)--> scelgo valori di l per i quali ho che M%l==0--> N blocchi numero intero 
cont2=0
for l in range(10,5001):
    if M%l==0:
        L[cont2]=l
        cont2+=1
print("Valori di L usati:")
print(L)

#Salvo le deviazioni standard calcolate con il metodo della media a blocchi
for i in range(cont):
        l=int(L[i])
        Gas_errU[i]=devstandard(M,l,gas[:,0])
        Gas_errP[i]=devstandard(M,l,gas[:,1])
        Liq_errU[i]=devstandard(M,l,liq[:,0])
        Liq_errP[i]=devstandard(M,l,liq[:,1])
        Sol_errU[i]=devstandard(M,l,sol[:,0])
        Sol_errP[i]=devstandard(M,l,sol[:,1])

#GAS
plt.plot(L,Gas_errU,label="errore U/N")
plt.plot(L,Gas_errP,label="errore P")
plt.xlabel("L")
plt.ylabel("$\sigma$")
plt.title("Errore al variare della larghezza dei blocchi (Gas)")
plt.legend()
plt.grid(True)
plt.show()

#LIQUIDO
plt.plot(L,Liq_errU,label="errore U/N")
plt.plot(L,Liq_errP,label="errore P")
plt.xlabel("L")
plt.ylabel("$\sigma$")
plt.title("Errore al variare della larghezza dei blocchi (Liquido)")
plt.legend()
plt.grid(True)
plt.show()

#SOLIDO
plt.plot(L,Sol_errU,label="errore U/N")
plt.plot(L,Sol_errP,label="errore P")
plt.xlabel("L")
plt.ylabel("$\sigma$")
plt.title("Errore al variare della larghezza dei blocchi (Solido)")
plt.legend()
plt.grid(True)
plt.show()
    
```

Nei grafici (soprattutto per quanto riguarda il liquido e il solido, che il gas ha dati già di partenza tra di loro molto scorrelati) si vede bene come l'errore statistico inizialmente tende ad aumentare al crescere di L per poi stabilizzarsi continuando ad oscillare all'interno di un range di valori. Il punto in cui l'errore si stabilizza è comparabile al $\tau_c$ in cui decade la funzione di autocorrelazione, confermando che blocchi larghi poche centinaia di step diano una buona stima del valore delle grandezze U/N e P.

```python
def media_blocchi(i,L,r):
    sum=0
    for j in range(L):
        k=j+i*L
        sum+=r[k]
    return sum/L

def media_progressiva(i,media):
    sum_prog=0
    for j in range(i+1):
        sum_prog+=media[j]

    return sum_prog/(i+1)

def err_progressivo(i,media):
    sum_prog=0
    sum_prog2=0
    for j in range(i+1):
        sum_prog+=media[j]
        sum_prog2+=media[j]**2
    sum_prog/=(i+1)
    sum_prog2/=(i+1)
    if i==0:
        err=0
    else:
        err=math.sqrt((sum_prog2-sum_prog**2)/i)
    return err

 


gas=np.loadtxt("Gas.dat",skiprows=15000)
liq=np.loadtxt("Liquido.dat",skiprows=15000)
sol=np.loadtxt("Solido.dat",skiprows=15000)

M=int(len(gas[:,0]))
L=200
N=int(M/L)
mgen = np.zeros(N)      #media gas energia
mgp = np.zeros(N)       #media gas pressione
mlen = np.zeros(N)
mlp = np.zeros(N)
msen = np.zeros(N)
msp = np.zeros(N) 

for i in range(N):
    mgen[i]=media_blocchi(i,L,gas[:,0])
    mgp[i]=media_blocchi(i,L,gas[:,1])
    mlen[i]=media_blocchi(i,L,liq[:,0])
    mlp[i]=media_blocchi(i,L,liq[:,1])
    msen[i]=media_blocchi(i,L,sol[:,0])
    msp[i]=media_blocchi(i,L,sol[:,1])

media_enG=np.zeros(N)   #media energia gas progressiva sui blocchi
media_pG=np.zeros(N)    #media pressione gas progressiva sui blocchi
media_enL=np.zeros(N)
media_pL=np.zeros(N)
media_enS=np.zeros(N)
media_pS=np.zeros(N)

err_enG=np.zeros(N)   #errore energia gas progressiva sui blocchi
err_pG=np.zeros(N)    #errore pressione gas progressiva sui blocchi
err_enL=np.zeros(N)
err_pL=np.zeros(N)
err_enS=np.zeros(N)
err_pS=np.zeros(N)

for i in range(N):
    media_enG[i]=media_progressiva(i,mgen)  
    media_pG[i]= media_progressiva(i,mgp)   
    media_enL[i]=media_progressiva(i,mlen)
    media_pL[i]=media_progressiva(i,mlp)
    media_enS[i]=media_progressiva(i,msen)
    media_pS[i]=media_progressiva(i,msp)

    err_enG[i]=err_progressivo(i,mgen)  
    err_pG[i]= err_progressivo(i,mgp)   
    err_enL[i]=err_progressivo(i,mlen)
    err_pL[i]=err_progressivo(i,mlp)
    err_enS[i]=err_progressivo(i,msen)
    err_pS[i]=err_progressivo(i,msp)

x=np.arange(N)*L

#GAS
plt.errorbar(x,media_enG,err_enG,label="U/N")
plt.errorbar(x,media_pG,err_pG,label="P")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.title("Simulazione gas")
plt.legend()
plt.grid(True)
plt.show()

#LIQUIDO
plt.errorbar(x,media_enL,err_enL,label="U/N")
plt.errorbar(x,media_pL,err_pL,label="P")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.title("Simulazione liquido")
plt.legend()
plt.grid(True)
plt.show()

#SOLIDO
plt.errorbar(x,media_enS,err_enS,label="U/N")
plt.errorbar(x,media_pS,err_pS,label="P")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.title("Simulazione solido")
plt.legend()
plt.grid(True)
plt.show()
```
