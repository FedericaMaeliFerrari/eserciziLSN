# EXERCISE 2.1

L'esercizio richiede di calcolare il valore dell'integrale $$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$
Per farlo si richiede di usare metodi MonteCarlo. Nel primo caso utilizzanod una distribuzione uniforme $[0,1]$, nel secondo caso usando il metodo dell'importance sampling. Questo metodo prevede di utilizzare un'altra distribuzione uniforme per minimizzare la varianza. la nuova distribuzione di probabilità deve dunque approssimare la funzione nell'intervallo voluto e deve essere una funzione invertibile. 

Ho approssimato la funzione $ \frac{\pi}{2}\cos(\pi x/2)$ nell'intervallo $[0,1]$ con la retta $2(1-x)$. I numeri sono stati dunque generati secondo la funzione cumulativa $1-\sqrt{1-y}$, e l'integrale da valutare per questo caso sarà dunque $$I = \int_0^1 \frac{\frac{\pi}{2}\cos(\pi x/2)}{2(1-x)} dx = 1$$.

Come si evince dai grafici sottostanti, l'integrale valutato secondo l'importance sampling approssima meglio il valore dell'integrale, e risulta avere un'incertezza minore rispetto all'integrale valutato utilizzando una distribuzione uniforme.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

N=100
L=1000
m=np.loadtxt("Integrale.dat") #carico dal file Dati.dat i vettori
x=np.arange(N)*L
yatt=np.zeros(N)

#grafico valore dell'integrale con distribuzione uniforme
plt.subplot(2,1,1)
plt.plot(x,m[:,0]-1,label="uniforme")
plt.plot(x,m[:,2]-1,label="sampling")
plt.plot(x,yatt,label="valore atteso")
plt.xlabel('#throws')
plt.ylabel('integrale')
plt.title("Valore dell'integrale (distribuzione uniforme)")
plt.grid(True)
plt.legend()
plt.show()

#grafico andamento dell'errore per l'integrale con distribuzione uniforme
plt.subplot(2,1,2)
plt.plot(x,m[:,1],label="uniforme")
plt.plot(x,m[:,3],label="sampling")
plt.xlabel('#throws')
plt.ylabel('errore')
plt.title("Andamento dell'errore (distribuzione uniforme)")
plt.grid(True)
plt.legend()
plt.show()

print("Incertezza con metodo distribuzione uniforme:")
print(m[N-1,1])
print("Inncertezza con metodo importance sampling:")
print(m[N-1,3])
```
