# EXERCISE 2.2
L'esercizio prevede una simulazione di Random Walk in due diverse condizioni. La prima su un reticolo cubico, in cui il punto è vincolato a muoversi per ogni tempo discreto lungo una sola direzione $x$, $y$, $z$ per un passo di lunghezza uguale ad $a=1$, sia in avanti che indietro. La seconda nel continuo: il punto si può muovere ad ogni tempo discreto in una direzione random sempre con passo $a=1$ , ovvero si posiziona sulla superficie di una sfera centrata nell'origine del cammino (dunque bisogna generare in modo uniforme un angolo solido con $\theta \in [0, \pi]$ e $\phi \in [0, 2\pi]$.

Al contrario che nel caso del Random Walk sul reticolo cubico, in cui la probabilità di effettuare un passo in qualsiasi direzione e verso seguiva la distribuzione uniforme, per il Random Walk nel continuo gli angoli $\theta$ e $\phi$ sono sati genrati secondo la distribuzione seno, per poi ricondursi alle coordinate $x$, $y$, $z$ attraverso le coordinate sferiche.

Ho condotto $10^4$ simulazioni al variare del numero dei passi $i \in [0,100]$. Alla fine di ogni set di passi ho calcolato la distanza al quadrato raggiunta dal punto di origine, sommando in quadtratura le coordinate in cui è arrivato il punto ($x^2+y^2+z^2$), e ho dunque fatto la media delle distanze raggiunte. 

Ho plottato la distanza $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ raggiunta al variare del numero dei passi $N$ con la relativa incertezza per entrambi i Random Walk. Entrambe le funzioni sono visibilmente proporzionali a $\sqrt{N}$. Ho dunque fittato le due distribuzioni attraverso una generica funzione $kN^b$. La funzione che fitta entrambi i risultati risulta essere $a\sqrt{N}$ (i parametri ottenuti dal fit sono riportati sotto ad ogni grafico).


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit

def f(N,k,pot):
    return k*N**pot

rw=np.loadtxt("RandomWalk.dat")
x=np.arange(100)

#RANDOM WALK DISCRETO
plt.errorbar(x,rw[:,0],rw[:,1],label="random walk")
plt.xlabel("Numero di passi")
plt.ylabel("Varianza")
plt.title("Random Walk su un reticolo cubico")
p_opt, p_cov = curve_fit(f, x, rw[:,0])
plt.plot(x,f(x,p_opt[0],p_opt[1]),label="fit")
plt.legend()
plt.grid(True)
plt.show()
print("Parametri che ottimizzano la curva:[k potenza]")
print(p_opt)
print("Errore massimo valutato:")
print(rw[99,1])

#RANDOM WALK CONTINUO
plt.errorbar(x,rw[:,2],rw[:,3],label="random walk")
plt.xlabel("Numero di passi")
plt.ylabel("Varianza")
plt.title("Random Walk nel continuo")
p_opt, p_cov = curve_fit(f, x, rw[:,2])
plt.plot(x,f(x,p_opt[0],p_opt[1]),label="fit")
plt.legend()
plt.grid(True)
plt.show()
print("Parametri che ottimizzano la curva:[k potenza]")
print(p_opt)
print("Errore massimo valutato:")
print(rw[99,3])
```

In entrambi i metodi la varianza risulta dunque proporzionale a $a\sqrt{N}$.Questi risultati indicano il comportamento di diffusione $$\frac{\delta p(x,t)}{\delta t}= D \frac{\delta^2 p(x,t)}{\delta x^2}$$
La soluzione dell'equazione sopracitata è l'equazione di una gaussiana, con sigma $a^2N$ --> i random walk simulati seguono la distribuzione di probabilità binomiale e, per il teorema del limite centrale, facendo tendere a infinito, si ritrova la distribuzione di probabilità gaussiana definita dalla soluzione dell'equazione di diffusione.
