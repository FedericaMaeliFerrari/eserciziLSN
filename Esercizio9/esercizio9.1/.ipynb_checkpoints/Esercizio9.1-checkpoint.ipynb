# EXERCISE 9.1

Lo scopo dell'esercizio è trovare il percorso più breve che deve fare un mercante per passare in 32 città (visitando ogni città una volta sola) tornando a fine percorso al punto di partenza.
Per risolvere questo problema viene richiesto di utilizzare l'algoritmo genetico, algoritmo nato per far variare una popolazione di cromosomi utilizzando metodi simili alla "selezione naturale".
L'algoritmo implementato funzione nel modo seguente:

**COSTRUTTORE:**

1) Si generano casualmente le coordinate delle 32 città (in un caso vincolandole su una circonferenza, nell'altro iscrivendole all'interno di un quadrato).

2) Si generano 100 percorsi casuali. Dopo il primo percorso (che non è altro che l'ordine delle città così come sono state generate) gli altri 99 percorsi non sono altro che variazioni del primo, in cui sono stati scambiati ogni volta una coppia di indici secondo l'operatore di Permutazione Pari (spiegato nel capitolo MUTAZIONE).

3) Si calcola per ogni percorso la lunghezza totale, considerando il tratto che servirebbe per tornare dall'ultima città sulla lista al punto di partenza

4) il vettore delle distanze L(x) calcolato al punto 3 viene riordinato in ordine crescente, e i 100 percorsi di conseguenza

**MUTAZIONE:**

Ogni vettore viene sottoposto a 4 mutazioni. Ogni mutazione ha il $10%$ di probabilità di presentarsi. Le mutazioni utilizzate sono:

1) Permutazione Pari--> già usata per generare i 100 percorsi all'interno del costruttore, vengono scambiati tra loro due valori consecutivi in una punto casuale del vettore

2) Shift--> vengono shiftate di n posizioni m città

3) Permutazione contigua--> vengono permutate m (m<N/2, N numero città totale) città con le altre m città contigue.

4) Inversione-->vengono invertite m città del vettore

Alla fine delle mutazioni controllo se le condizioni al contorno sono ancora rispettate (ovvero se ogni città appare solo una volta all'interno del vettoree se la città iniziale è sempre la stessa, ovvero la prima generata (indice 0)) per mezzo dell'operatore check. Se non viene rilevato nessun errore i percorsi vengono di seguito ordinati in ordine crescente secondo la distanza e si procede con la parte terza parte del programma.

**NEW GENERATION:**

Viene creata la nuova popolazione per mezzo dell'operatore crossover,che ha prob >$50%$ di presentarsi ( io ho fatto $70%$ di probabilità). Questo operatore funziona nel modo seguente:

1) sceglie due vettori (percorsi) dalla popolazione vecchia

2) li "taglia" virtualmente nella stessa posizione

3) riordina gli elementi tagliati in ciascun vettore secondo l'ordine in cui appaiono nel vettore compagno

4) I due vettori così crossoverati entrano a far parte della nuova popolazione

I vettori selezionati per il crossover vengono scelti tramite l'operatore Select che favorisce la scelta dei percorsi con lunghezza del perocorso L(x) minore-->viene selezionato il percorso j-esimo con $j=N\times int(r^2)$, con r numero generato casualmente tra $0$ e $1$. 
Nel caso non avvenisse il crossover i percorsi selezionati entrano direttamente a far parte della nuova popolazione. Generata tutta la popolazione anche su di essa viene fatto un check e viene riordinata secondo L(x).


Eseguendo gli ultimi due passaggi più volte si ottengono ogni volta delle nuove popolazioni di percorsi con lunghezza complessiva L(x) più piccola della popolazione precedente.Iterando il processo abbastanza a lungo il sistema convergerà al percorso che minimizza la lunghezza totale che dovrà percorrere il mercante.

Ho condotto la simulazione nei due casi (città su circonferenza/città iscritte nel quadrato) per $100$ blocchi, ognuno composto da $1000$ step. In figura sono riportati gli andamenti della lunghezza $L(x)$ del percorso più breve dopo ogni crossover e della lunghezza media $\langle L(x)\rangle$ di tutti i percorsi al variare del numero di blocchi. Per far ciò ho utilizzato il codice della media a blocchi simile a quelli utilizzati nelle esercitazioni precedenti.
Si può notare, oltre a una diminuzione di $L(x)$ e di  $\langle L(x)\rangle$, anche una diminuzione dell'errore associato alla misura, in quanto con l'aumentare dei lanci dell'algoritmo i percorsi tendono appunto più a convergere.
Ho inoltre riportato un disegno del percorso che dovrà fare il mercante per attraversare le 32 città nei due casi, ovvero il percorso che al termine della simulazione minimizza la distanza da percorrere.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math


pathCirc=np.loadtxt("bestpath_circ.dat")
coord=np.loadtxt("Città_circonferenza.dat")
Lx_min=np.loadtxt("Lx_circ.dat")
Lx_media=np.loadtxt("Lx_media_circ.dat")

plt.errorbar(Lx_min[:,0],Lx_min[:,2],Lx_min[:,3])
plt.xlabel("N_blocchi")
plt.ylabel("L(x)")
plt.title("stima di L(x) best path")
plt.grid(True)
plt.show()

plt.errorbar(Lx_media[:,0],Lx_media[:,2],Lx_media[:,3])
plt.xlabel("N_blocchi")
plt.ylabel("L(x)")
plt.title("stima di L(x)media")
plt.grid(True)
plt.show()

#creo i vettori ordinati che minimizzano il percorso L(x) seguendo gli indici caricati in pathCirc(ordino x e y per poterli graficare)
vettX=np.ones(33)
vettY=np.ones(33)

for i in range(32):
    j=int(pathCirc[i])
    vettX[i]=coord[j,1]
    vettY[i]=coord[j,2]
vettX[32]=vettX[0]          #chiudo il vettore facendolo tornare alla città iniziale
vettY[32]=vettY[0]

plt.plot(coord[:,1],coord[:,2],"ro")
plt.plot(vettX,vettY) 
plt.xlabel("x")
plt.ylabel("y")
plt.title("percorso che minimizza L(x)")
plt.grid(True)
plt.show()
```

```python
pathQuad=np.loadtxt("bestpath_quadr.dat")
coord_quad=np.loadtxt("Città_quadrato.dat")
Lx_min_quad=np.loadtxt("Lx_quadr.dat")
Lx_media_quad=np.loadtxt("Lx_media_quadr.dat")

plt.errorbar(Lx_min_quad[:,0],Lx_min_quad[:,2],Lx_min_quad[:,3])
plt.xlabel("N_blocchi")
plt.ylabel("L(x)")
plt.title("stima di L(x) best path")
plt.grid(True)
plt.show()

plt.errorbar(Lx_media_quad[:,0],Lx_media_quad[:,2],Lx_media_quad[:,3])
plt.xlabel("N_blocchi")
plt.ylabel("L(x)")
plt.title("stima di L(x)media")
plt.grid(True)
plt.show()

#creo i vettori ordinati che minimizzano il percorso L(x) seguendo gli indici caricati in pathCirc(ordino x e y per poterli graficare)
vettX_q=np.ones(33)
vettY_q=np.ones(33)

for i in range(32):
    j=int(pathQuad[i])
    vettX_q[i]=coord_quad[j,1]
    vettY_q[i]=coord_quad[j,2]
vettX_q[32]=vettX_q[0]           #chiudo il vettore facendolo tornare alla città iniziale
vettY_q[32]=vettY_q[0]

plt.plot(coord_quad[:,1],coord_quad[:,2],"ro")
plt.plot([0,2,2,0,0],[0,0,2,2,0],"b--")
plt.plot(vettX_q,vettY_q) 
plt.xlabel("x")
plt.ylabel("y")
plt.title("percorso che minimizza L(x)")
plt.grid(True)
plt.show()
```
